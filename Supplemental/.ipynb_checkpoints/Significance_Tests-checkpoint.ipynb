```python
import pandas as pd
from scipy.stats import chi2_contingency

df = pd.read_csv('participation_ratios.csv', index_col=1)
df.drop(columns=['Unnamed: 0'], inplace=True)

df.head()
```

```python
checks = []
ratios = ['Native American Participation Ratio',
               'Pacific Islander Participation Ratio',
               'Multiracial Participation Ratio']
for ratio in ratios:
    states = list(df[df[ratio]>1]['State'])
    states = [(state, ratio) for state in states]
    checks += states

checks
```

```python
for check in checks:
    group = check[1].replace(' Participation Ratio', '')
    a = df.loc[check[0], (group + ' Number')]
    b = df.loc[check[0], (group + ' Number AP')]
    c = df.loc[check[0], 'Total']
    d = df.loc[check[0], 'Total AP']
    chi2 = chi2_contingency([[a,b],[c,d]])
    if chi2[1] < 0.1:
        print(f'Significant result {check} with p-value {chi2[1]}')
        print(
            f'There are {a} {group} students enrolled ({(a/c*100):.2f}% of all students), '
            f'and there are {b} {group} test takers ({b/d*100:.2f}%).\n'
        )
```
